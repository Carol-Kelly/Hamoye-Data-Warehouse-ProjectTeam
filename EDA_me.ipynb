```python
#import important libraries for plotting and visualizations
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
sns.set_style('darkgrid')
```

```python
#with the new subsets created, let's make a line plot first to compare a particular job title between the three genders
plt.figure(figsize= (16, 10))
plt.plot(df_females_['female managers'], color = 'red', label = '% Salary for female managers', marker = '+');
plt.plot(df_males_['male managers'], color = 'blue',label = '% Salary for male managers', marker = '*' );
plt.plot(df_persons_['person managers'], color = 'green', label = '% Salary for person managers', marker = '*');
plt.xlabel("Year",size= 12)
plt.ylabel("Percentage of salary",size=12)
plt.title("Comparison of salary of managers among the genders",size=14)
plt.legend();
```

- There is an upward trend in the percentage of salaries as the years go by with a sharp decrease from 2007 after which it slowly began to increase. This could be as a result of the global economic crises.
[https://www.britannica.com/event/financial-crisis-of-2007-2008/Effects-and-aftermath-of-the-crisis](https://)
- From the plot above, we can see that the male managers have always been earning more than both female and persons' genders.
- There is a peak in 2007 and 2014 respectively, unfortunateky there is isn't enough data to find out why.

```python
df_females_rename = df_females_.rename(columns = {'females' : 'Gender', 'female managers' : 'managers', 'female professionals' : 'professionals',
       'female technicians and trades workers' : 'technicians and trades workers',
       'female community and personal service workers' : 'community and personal service workers',
       'female clerical and administrative workers' : 'clerical and administrative workers', 'female sales workers' : 'sales workers',
       'female machinery operators and drivers' : 'machinery operators and drivers', 'female labourers' : 'labourers'})

df_males_rename = df_males_.rename(columns = {'males' : 'Gender', 'male managers' : 'managers', 'male professionals' : 'professionals',
       'male technicians and trades workers' : 'technicians and trades workers',
       'male community and personal service workers' : 'community and personal service workers',
       'male clerical and administrative workers' : 'clerical and administrative workers', 'male sales workers' : 'sales workers',
       'male machinery operators and drivers' : 'machinery operators and drivers', 'male labourers' : 'labourers'})

df_persons_rename = df_persons_.rename(columns = {'persons' : 'Gender', 'person managers' : 'managers', 'person professionals' : 'professionals',
       'person technicians and trades workers' : 'technicians and trades workers',
       'person community and personal service workers' : 'community and personal service workers',
       'person clerical and administrative workers' : 'clerical and administrative workers', 'person sales workers' : 'sales workers',
       'person machinery operators and drivers' : 'machinery operators and drivers', 'person labourers' : 'labourers'})
```

- Since it might take time to plot for each of the job titles, we might as well make a new dataset extracting out the genders to form a "Gender" column such that each job title for each column also has a corresponding gender.

```python
df_males_rename
```

```python
df_full = pd.concat([df_females_rename, df_males_rename, df_persons_rename], axis = 0)
```

```python
df_full.head()
```

# Explanatory Data Analysis

Univariate plots

```python
df_full.hist(figsize=(16,10));
```

- The plot above shows distribution of the different job titles and distribution of the percentages of their salaries irrespective of gender.
- The 'managers' and  'professionals' seem to be the categories with higher payment in salaries .
- Majority of the managers earn within the range of 35-40 . 

```python
sns.countplot(df_full['Gender']);
plt.title("Gender Distribution")
```

- It is likely to have that gender count is equal. All being 14, which  is originally the initial number of rows before preprocessing.


Bivariate and Multi-variate analysis

```python
fig1 = sns.FacetGrid(df_full ,hue = 'Gender', aspect = 5)
fig1.map(sns.kdeplot, 'managers', shade = True)
oldest = df_full['managers'].max()
fig1.set(xlim=(0,oldest))
fig1.add_legend()
```

From the 'KDE' plot above comparing by gender the salaries of managers, the :
- The minimum salary for female managers is less than that of male and persons.
- The mean salary of male managers is ~42 . 
- Only few managers of the 3 genders earn above 50 . 
- We also note that 'Persons' as managers earn quite high.

```python
pd.DataFrame(df_full.groupby('Gender')['managers'].mean()).plot(kind='bar', color= 'brown',alpha=0.5);
```

- The plot above gives summary of inferences stated from the KDE plot above.

```python
#extracting the dataset gender-wise
df_sub_female = df_full[df_full['Gender']=='Females']
df_sub_male = df_full[df_full['Gender']=='Males']
df_sub_persons = df_full[df_full['Gender']=='Persons']
```

```python
df_sub_female.head()
#df_sub_male.head()
#df_sub_persons.head() Let's check and un for females only 
```

```python
df_sub_female.plot(linewidth = 3, figsize = (20,9));
plt.xlabel('Year')
plt.ylabel('% Salary')
plt.title('Year vs % Salary of Females');
```

- The plot above shows the comparison of the job categories of females over the years and the percentage of salary earned. 
- There's a positive gradual increase in the salaries of the job categories.
- At the year 2008, the salaries of 'professionals' surpassed that of 'managers' and the trend has been that way. 
- The 'labourers' and 'sales workers' have the lowest earnings. 
- It would be nice to compare the earnings of the 'professionals' by gender .

```python
df_sub_male.plot(linewidth = 3, figsize = (20, 10));
plt.xlabel('Year')
plt.ylabel('% Salary')
plt.title('Year vs % Salary of Males');
```

- Unlike the female professionals, the male managers and professionals have an intertwined relationship where by they might either have same earnings (2007,2010,2013,2016) or the 'professionals' earning more than 'managers' or otherwise (2009).
- Similarly to the females, the male 'labourers' and 'sales workers' have the lowest earnings among all the job categories. 

```python
df_sub_persons.plot(linewidth = 3, figsize = (20, 10));
plt.xlabel('Year')
plt.ylabel('% Salary')
plt.title('Year vs % Salary of Persons');
```

- Just like 'histogram' distribution from earlier, the 'managers' and 'professionals' earn much higher salaries than other job categories. This is of course expected as there are different levels in the labor market and of which some are higher in demand than the others.
