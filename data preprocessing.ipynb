We restructure the data into three separate datasets consisting of the females, males and persons each.

```python
df = df.rename(columns={'year':'job_title'})
df
```

```python
df.set_index('job_title', inplace=True)
total_df = df.T
total_df = total_df.drop(['females','males','persons'], axis=0)
total_df
```

Females Average hourly earning dataset 

```python
#Extract the females features only from the datasets
df_females = total_df[0:8]
df_females
```

Males Average hourly earning dataset

Persons Average hourly earning dataset

```python
#Extract the persons features only from the datasets
df_persons = total_df[16:]
df_persons
```
