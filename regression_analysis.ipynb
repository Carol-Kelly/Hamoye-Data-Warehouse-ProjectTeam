```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Preshynodis/Hamoye-Data-Warehouse-ProjectTeam/main/earning.csv", delimiter=';', nrows=None)

df.shape
```

```python
df
```

```python
# Getting the columns of male and female with respect to their jobs in a single dataframe 

df_managers = df.iloc[0:, [0, 2, 11]]
df_professionals = df.iloc[:, [0, 3, 12]]
df_techniciansandtradesworkers = df.iloc[:, [0, 4, 13]]
df_communityandpersonalserviceworkers = df.iloc[:, [0, 5, 14]]
df_clericalandadministrativeworkers = df.iloc[:, [0, 6, 15]]
df_salesworkers = df.iloc[:, [0, 7, 16]]
df_machineryoperatorsanddrivers = df.iloc[:, [0, 8, 17]]
df_labourers = df.iloc[:, [0, 9, 18]]
```

```python
df_labourers
```

```python
pd.options.mode.chained_assignment = None

# From the newly created dataframe find the differnce between the make and the female ( Earnings ) with respect to their jobs. 

df_managers['managers_diff'] = df_managers['malemanagers'] - df_managers['femalesmanagers']
df_professionals['proffesionals_diff'] = df_professionals['maleprofessionals'] - df_professionals['femaleprofessionals']
df_techniciansandtradesworkers['techniciansandtradesworkers_diff'] = df_techniciansandtradesworkers['maletechniciansandtradesworkers'] - df_techniciansandtradesworkers['femaletechniciansandtradesworkers']
df_communityandpersonalserviceworkers['communityandpersonalserviceworkers_diff'] = df_communityandpersonalserviceworkers['malecommunityandpersonalserviceworkers'] - df_communityandpersonalserviceworkers['femalecommunityandpersonalserviceworkers']
df_clericalandadministrativeworkers['clericalandadministrativeworkers_diff'] = df_clericalandadministrativeworkers['maleclericalandadministrativeworkers'] - df_clericalandadministrativeworkers['femaleclericalandadministrativeworkers']
df_salesworkers['salesworkers_diff'] = df_salesworkers['malesalesworkers'] - df_salesworkers['femalesalesworkers']
df_machineryoperatorsanddrivers['machineryoperatorsanddrivers_diff'] = df_machineryoperatorsanddrivers['malemachineryoperatorsanddrivers'] - df_machineryoperatorsanddrivers['femalemachineryoperatorsanddrivers']
df_labourers['labourers_diff'] = df_labourers['malelabourers'] - df_labourers['femalelabourers']

```

```python
df_machineryoperatorsanddrivers
```

```python
df_managers
```

```python
sns.set_style('white')

# seaborn plot to find the relationship between the difference in earnings with respecet to the years 

sns.lmplot(x ="year", y ="managers_diff", data = df_managers, line_kws= {'lw' : 4,'color' : 'red'})
sns.lmplot(x ="year", y ="proffesionals_diff", data = df_professionals, line_kws= {'lw' : 4,'color' : 'Black'})
sns.lmplot(x ="year", y ="techniciansandtradesworkers_diff", data = df_techniciansandtradesworkers, line_kws= {'lw' : 4,'color' : 'yellow'})
sns.lmplot(x ="year", y ="communityandpersonalserviceworkers_diff", data = df_communityandpersonalserviceworkers, line_kws= {'lw' : 4,'color' : 'orange'})
sns.lmplot(x ="year", y ="clericalandadministrativeworkers_diff", data = df_clericalandadministrativeworkers, line_kws= {'lw' : 4,'color' : 'purple'})
sns.lmplot(x ="year", y ="salesworkers_diff", data = df_salesworkers, line_kws= {'lw' : 4,'color' : 'blue'})
sns.lmplot(x ="year", y ="machineryoperatorsanddrivers_diff", data = df_machineryoperatorsanddrivers, line_kws= {'lw' : 4,'color' : 'pink'})
sns.lmplot(x ="year", y ="labourers_diff", data = df_labourers, line_kws= {'lw' : 4,'color' : 'green'})

```



The plots above are showing the relationship between the male and female earning with respect to job title. In the relationship, the differnce for most jobs are positive meaning that the male earns more that the female while few it shows the opposite
