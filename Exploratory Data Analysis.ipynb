Females Average hourly earning 

```python
#Extract the females features only from the datasets
df_females = total_category[0:9]
df_females=df_females.drop('females', axis=0)
df_females
```

```python
#Extract the male features only from the datasets
df_males = total_category[9:18]
df_males=df_males.drop('males', axis=0)
df_males
```

Persons Average hourly earning 

```python
#Extract the persons features only from the datasets
df_persons = total_category[18:]
df_persons=df_persons.drop('persons', axis=0)
df_persons
```

# Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(12,8))
plt.xlabel('Year')
plt.ylabel('Earnings')
plt.title('Female Earnings Per Year')
leg_list=['Managers', 'Professionals','Technicians and Trades Workers','Community and Personal Service Workers','Clerical and Administrative Workers', 'Sales Workers','Machinery Operators and Drivers', 'Labourers']
x=df_females.columns
y0=df_females.iloc[0]
y1=df_females.iloc[1]
y2=df_females.iloc[2]
y3=df_females.iloc[3]
y4=df_females.iloc[4]
y5=df_females.iloc[5]
y6=df_females.iloc[6]
y7=df_females.iloc[7]
plt.plot(x,y0,marker='o')
plt.plot(x,y1,marker='o')
plt.plot(x,y2,marker='o')
plt.plot(x,y3,marker='o')
plt.plot(x,y4,marker='o')
plt.plot(x,y5,marker='o')
plt.plot(x,y6,marker='o')
plt.plot(x,y7,marker='o')
plt.legend(leg_list)
plt.savefig('Females Earnings per Year')
plt.show()
```

Results from the Exploratory Data Analysis:

- Generally, earnings increased regardless of the gender or the type of job over the years.

- Also in 2016, male and female machinery operators and drivers approximately earned the same.
- In 2007, female managers had a sharp increase in their earnings
- In 2014, female technicians and trade workers had a sharp increase in their earnings
- In 2008, male community and personal service workers had a sharp increase in their earnings
- In 2005, female machinery operators and drivers had a sharp increase in their earnings followed by an increased in 2014 and 2015
