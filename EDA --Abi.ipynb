### EDA

```python
import numpy as np
total_df = total_df.T.reset_index().rename(columns={"job_title":"year"})
```

```python
total_df
```

```python
female = total_df[total_df.columns[0:9]]
female.rename(columns={'female_managers':'Managers', 'female_professionals':"Professionals",
       'female_technicians_and_trades_workers':"Technicians and Trade workers",
       'female_community_and_personal_service_workers':"Community and Personal Service workers",
       'female_clerical_and_administrative_workers':"Clerical and Administrative workers", 'female_sales_workers':"Sales workers",
       'female_machinery_operators_and_drivers':"Machinery Operators and Drivers", 'female_labourers':"Labourers"},inplace=True)
```

```python
#rearranging the female subset
list_df = [] #empty list
for year in female["year"]: #iterating through the year column of the female subset
    x = female[female["year"] == year].T.reset_index() #transposing and resetting the index
    x["year"] = year #creating a new column containing year values
    x["gender"] = 'Female' #creating a new column containing gender values
    x.drop(index=x.index[0],axis=0,inplace=True) #dropping the first two rows which are the old year and gender columns that were equally transposed
    list_df.append(x) #appending the resulting dataframe to the external list
for df in list_df: #changing the column names for each dataframe in the list
    df.columns = ["Job_title","earnings","year","gender"]
f = pd.concat(list_df) #concatenating the dataframes in the list together
f.reset_index(drop=True,inplace=True) #resetting the index
f #the female earnings dataframe
```

```python
male = total_df.iloc[:,np.r_[0,9:17]]
male.rename(columns={'male_managers':'Managers', 'male_professionals':"Professionals",
       'male_technicians_and_trades_workers':"Technicians and Trade workers",
       'male_community_and_personal_service_workers':"Community and Personal Service workers",
       'male_clerical_and_administrative_workers':"Clerical and Administrative workers", 'male_sales_workers':"Sales workers",
       'male_machinery_operators_and_drivers':"Machinery Operators and Drivers", 'male_labourers':"Labourers"},inplace=True)
```

```python
#rearranging the male subset
list_df = [] #empty list
for year in male["year"]: #iterating through the year column of the male subset
    x = male[male["year"] == year].T.reset_index() #transposing and resetting the index
    x["year"] = year #creating a new column containing year values
    x["gender"] = 'Male' #creating a new column containing gender values
    x.drop(index=x.index[0],axis=0,inplace=True) #dropping the first two rows which are the old year and gender columns that were equally transposed
    list_df.append(x) #appending the resulting dataframe to the external list
for df in list_df: #changing the column names for each dataframe in the list
    df.columns = ["Job_title","earnings","year","gender"]
m = pd.concat(list_df) #concatenating the dataframes in the list together
m.reset_index(drop=True,inplace=True) #resetting the index
m #the male earnings dataframe
```

```python
persons = total_df.iloc[:,np.r_[0,17:25]]
persons.rename(columns={'person_managers':'Managers', 'person_professionals':"Professionals",
       'person_technicians_and_trades_workers':"Technicians and Trade workers",
       'person_community_and_personal_service_workers':"Community and Personal Service workers",
       'person_clerical_and_administrative_workers':"Clerical and Administrative workers", 'person_sales_workers':"Sales workers",
       'person_machinery_operators_and_drivers':"Machinery Operators and Drivers", 'person_labourers':"Labourers"},inplace=True)
```

```python
#rearranging the persons subset
list_df = [] #empty list
for year in persons["year"]: #iterating through the year column of the persons subset
    x = persons[persons["year"] == year].T.reset_index() #transposing and resetting the index
    x["year"] = year #creating a new column containing year values
    x["gender"] = 'Persons' #creating a new column containing gender values
    x.drop(index=x.index[0],axis=0,inplace=True) #dropping the first two rows which are the old year and gender columns that were equally transposed
    list_df.append(x) #appending the resulting dataframe to the external list
for df in list_df: #changing the column names for each dataframe in the list
    df.columns = ["Job_title","earnings","year","gender"]
p = pd.concat(list_df) #concatenating the dataframes in the list together
p.reset_index(drop=True,inplace=True) #resetting the index
p #the persons earnings dataframe
```

```python
data = pd.concat([f,m,p]).sort_values(by='year').reset_index(drop=True)
data
```

### Are there jobs where females earn higher than males?

```python
#comparing male earnings and female earnings
m["comparison"] = np.where(m["earnings"] > f["earnings"], 'Higher', 'Lower')
m[m["comparison"] == 'Lower'] #checking for jobs where male earnings were lower than female earnings.
```

#### There were three cases where male earned lower than females.
#### In 2005, male machinery operators and drivers earned (on an average) lower than their female co-workers.
#### The same occurred in 2015 and 2016.

```python
#creating a new column that shows the difference between male earnings and female earnings.
m["female_earnings"] = f["earnings"] #adding female earnings to male dataframe
m["difference"] = m["earnings"] - m["female_earnings"] #computing the difference and storing in a new column
m.sort_values("difference") #sorting the dataframe by the difference column in ascending order
```

#### The negative signs show that truly females earned more than the males in those three cases (machinery operators and drivers in 2015,2005 and 2016).
#### The difference was higher in 2015 and least in 2016.
#### In short, out of the 112 different cases, females earned higher than males only three times.

```python
#checking for cases where males and females earned in equal rates.
m[m["difference"] == 0]
```

#### Males earned higher in other cases

### Which jobs had the highest difference in male and female earnings?

```python
pl = m.sort_values("difference",ascending=False)[:10]
sns.barplot(y="Job_title",x="difference",data=pl)
```
