#### EDA ( EXPLANATORY DATA ANALYSIS )
It should be noted that 'Persons' in the dataset are employees who did not specify their gender (whether they are male or female)

Let's plot the trend of average hourly earnings per gender.

```python
#define a user-defined function that plots the gender earnings from year 2004-2017 per job titles
def plot_gender(df_gender, gender):
  plt.figure(figsize=(12,8))
  plt.xlabel('Year')
  plt.ylabel('Earnings')
  plt.title(f'{gender} Earnings Per Year in Australia')
  leg_list=['Managers', 'Professionals','Technicians and Trades Workers','Community and Personal Service Workers','Clerical and Administrative Workers', 'Sales Workers','Machinery Operators and Drivers', 'Labourers']
  x=df_gender.columns
  for i in range(8):
    y=df_gender.iloc[i]
    plt.plot(x,y,marker='o')
  
  plt.legend(leg_list)
  #  plt.savefig('Females Earnings per Year')
  plt.show()
```

```python
# call the plot_gender function defined 
plot_gender(df_females, 'Females')
plot_gender(df_males, 'Males')
plot_gender(df_persons, 'Persons')
```

Generally as seen above, earnings increased regardless of the gender or the type of job over the years.

Visualization that compares males, females and persons per each job titles.

```python
#create a role variable that stores the job titles 
role = df_males.index.tolist()
role = [y.replace("male_", "") for y in role]
for i in range(8):  
  plt.figure(figsize=(12,7))
  plt.xlabel('Year')
  plt.ylabel('Earnings')
  title = 'Male vs Female vs Person '+str(role[i])+' Earnings Per Year in Australia'
  plt.title(title)
  x=total_df.columns
  y1=df_females.iloc[i]
  y2=df_males.iloc[i]
  y3=df_persons.iloc[i]
  list = [y1,y2,y3]
  for j in list:
    plt.plot(x,j, marker='o')
    plt.legend(['Females', 'Males','Persons'])
  # plt.savefig('Males vs Females Managers')
  plt.show()
```

Results from the Exploratory Data Analysis:
- Managers and Professionals earn more than other types of job over the years regardless of the gender
- Males earn more than females over the years regardless of the type of job except female machinery operators and drivers in 2005 and 2015 when they earned higher than their counterparts.
- Also in 2016, male and female machinery operators and drivers approximately earned the same.
- In 2007, female managers had a sharp increase in their earnings
- In 2014, female technicians and trade workers had a sharp increase in their earnings
- In 2008, male community and personal service workers had a sharp increase in their earnings
- In 2005, female machinery operators and drivers had a sharp increase in their earnings and then an increased in 2014 and 2015
- In 2005, 2015, 2016, we can see that there is an unusual occurence that shows the female earning more in the Machinery Operator and Driver Job title. 

```python
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline 

df_females2 = df_females.copy(deep=False).T
df_males2 = df_males.copy(deep=False).T
df_persons2 = df_persons.copy(deep=False).T

df_females2.columns = [c.replace("female_", "") for c in (df_females2.columns)]
df_females2['gender']='female'
df_males2.columns   = [c.replace("male_", "") for c in (df_males2.columns)]
df_males2['gender']='male'
df_persons2.columns = [c.replace("person_", "") for c in (df_persons2.columns)]
df_persons2['gender']='person'

df_full = pd.concat([df_females2, df_males2, df_persons2], axis = 0)

df_females2.columns = df_females2.columns.str.replace('female_', '')
df_males2.columns = df_males2.columns.str.replace('male_', '')
df_persons2.columns = df_persons2.columns.str.replace('person_', '')

df_full = pd.concat([df_females2, df_males2, df_persons2], axis = 0)

df_full2 = df_full.drop("gender",axis=1)

for i in df_full2.columns:
  df_full2[i] = df_full2[i].apply(int)

df_full2.hist(figsize=(16,10));
```

- The plot above shows distribution of the different job titles and distribution of the percentages of their salaries irrespective of gender.
- The 'managers' and 'professionals' seem to be the categories with higher payment in salaries .
- Majority of the managers earn within the range of 35-40 Australian dollars (AUD) . 

```python
# Bivariate and Multi-variate Analysis 

fig1 = sns.FacetGrid(df_full,hue = 'gender', aspect = 5)
fig1.map(sns.kdeplot, 'managers', shade = True)
oldest = df_full['managers'].max()
fig1.set(xlim=(0,oldest))
fig1.add_legend()
```

From the 'KDE' plot above comparing by gender the salaries of managers, the :
- The minimum salary for female managers is less than that of male and persons.
- The mean salary of male managers is about 43 Australian dollars (AUD).
- The mean salary of female managers is about AUD35.
- The mean salary of persons managers is about AUD40 
- Only few managers of the 3 genders earn above AUD50 . 

```python
job_types = df_full.columns
df_full3 = df_full.reset_index()
df_full3 = df_full3.melt(id_vars=['job_title', 'gender'], value_vars= job_types,var_name= 'career_type', value_name='earning_per_hour' )

career = df_full3.groupby(['career_type', 'gender'])['earning_per_hour'].mean()
career= career.reset_index()

plt.figure(figsize=(10, 8))
sns.barplot(data = career, x = 'earning_per_hour', y = 'career_type', hue = 'gender')
plt.ylabel('Career Type')
plt.xlabel('Earnings per hour')
plt.title('Average Hourly Earnings for Different Career in Australia')
plt.show()
```

The Plot shows how hourly wages differ across the various job title

It can be observed from the figure that irrespective of cadre, the male gender has the highest hourly pay while the female has the least. Also the difference between male and the other genders tends to increase for job titles with more hourly pay.

It is observed that the managers and the professionals are the most paid cadres.
