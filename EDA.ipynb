The general earnings trend remained the same throughout the period across all categories

Managers and Professionals earned the most, with professionals having a slight edge

This was followed by (clerical and administrative workers), (technicians and trades workers), (machinery operators and drivers) and (community and personal service workers)

The labourers were the least earning group irrespective of all genders
