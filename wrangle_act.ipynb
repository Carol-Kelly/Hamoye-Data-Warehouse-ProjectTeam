### Data Cleaning


```python
df_clean = df.copy()
df_clean.info()
```

```python
# Separate the dataset into male_df, female_df and person_df
df_clean.set_index('year', inplace=True)
total_category = df_clean.T
total_category
```

```python
#Extract the male features only from the datasets
df_males = total_category[9:18]
df_males
```

```python
males_df = df_males.T
males_df.reset_index(inplace = True)
males_df.head()
```

```python
males_df.rename(columns ={ 'males':'gender'}, inplace=True)
```

```python
columns_list = []
for column in males_df.columns:
    if 'male' not in column:
        columns_list.append(column)
    else:
        new = column.replace('male', '')
        columns_list.append(new)
print(columns_list)
```

```python
males_df.columns = columns_list
males_df.head()
```

```python
#Extract the females features only from the datasets
df_females = total_category[0:9]
#df_females=df_females.drop('females', axis=0)
df_females
```

```python
females_df = df_females.T
females_df.reset_index(inplace = True)
females_df.head()
```

```python
females_df.columns = columns_list
```

```python
females_df.head()
```

```python
#Extract the persons features only from the datasets
df_persons = total_category[18:]
#df_persons=df_persons.drop('persons', axis=0)
df_persons
```

```python
persons_df = df_persons.T
persons_df.reset_index(inplace = True)
persons_df.head()
```

```python
persons_df.columns = columns_list
persons_df.head()
```

```python
# concatenate the dataframes to one
df_clean = pd.concat([males_df, females_df, persons_df], axis =0)
df_clean.shape
```

```python
df_clean.columns
```

```python
df_clean = df_clean.melt(id_vars=['year', 'gender'], value_vars= [ 'managers', 'professionals','techniciansandtradesworkers', 'communityandpersonalserviceworkers','clericalandadministrativeworkers', 'salesworkers','machineryoperatorsanddrivers', 'labourers'],var_name= 'career_type', value_name='earning_per_hour' )
df_clean.shape
```

```python
df_clean.head()
```

```python
df_clean.info()
```

### Analysis

#### 1. What is the average hourly earning of the various gender for total period of analysis(2004 - 2017)?

```python
total_gender = df_clean.groupby('gender')['earning_per_hour'].mean()
total_gender = total_gender.reset_index()
total_gender.head()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data = total_gender, x= 'gender', y = 'earning_per_hour')
plt.xlabel('Gender')
plt.ylabel('Earnings per hour')
plt.title('Average Hourly Earnings of the Various Gender')
```

The figure shows the average houly early of the 8 career types over 14years period (2004 - 2017) for the various gender. The male gender receives the highest houly wage of AUD30.59 while the female gender receives the least of AUD26.05.

```python
yearly_gender = df_clean.groupby(['year', 'gender'])['earning_per_hour'].mean()
yearly_gender = yearly_gender.reset_index()
yearly_gender.head()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data = yearly_gender, x = 'year', y = 'earning_per_hour', hue= 'gender')
plt.xlabel('Year')
plt.ylabel('Earnings per hour')
plt.title('Yearly Average Hourly Earnings of the Various Gender')
```

The graph reveals that there is a constant rising trend of houly wages for the various gender. The difference in hourly wages for the various gender has remain relatively constant over the years.

#### 2. How does gender hourly wages differ across the various career?

```python
career = df_clean.groupby(['career_type', 'gender'])['earning_per_hour'].mean()
career= career.reset_index()
career.head()
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(data = career, x = 'earning_per_hour', y = 'career_type', hue = 'gender')
plt.ylabel('Career Type')
plt.xlabel('Earnings per hour')
plt.title('Average Hourly Earnings for Different Career in Australia')
plt.show()
```

It can be observed from the figure that irrespective of career, the male gender has the highest hourly pay while the female has the least. Also the difference between male and the other genders tends to increase for careers with more hourly pay.
